```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time
```

```python
A1 = np.array([0,0,0])
A2 = np.array([0,0,1])
A3 = np.array([0,1,0])
A4 = np.array([1,0,0])
A5 = np.array([0,1,1])
A6 = np.array([1,0,1])
A7 = np.array([1,1,0])
A8 = np.array([1,1,1])
A9 = np.array([0.5,0.5,0.5])
A = np.array([A1,A2,A3,A4,A5,A6,A7,A8,A9])
```

```python
def initialize(n):
    X = np.linspace(0,1,n)
    Y = np.linspace(0,1,n)
    Z = np.linspace(0,1,n)
    side = X[1] - X[0]
    return (X,Y,Z, side)

def nearest_point(a):
    d = []
    for i in A:
        d.append(np.linalg.norm(a - i))
    m = np.argmin(d)
    return m

def find_ans(n):
    Answer_8 = []
    Answer_1 = []
    Answer_7 = []
    Answer_3 = []
    X,Y,Z,side = initialize(n)
    for i in X:
        for j in Y:
            for k in Z:
                x = i + side/2
                y = j + side/2
                z = k + side/2
                res = nearest_point(np.array([x,y,z]))
                if res==1:
                    Answer_1.append(np.array([x,y,z]))
                elif res==3:
                    Answer_3.append(np.array([x,y,z]))
                elif res==7:
                    Answer_7.append(np.array([x,y,z]))
                elif res==8:
                    Answer_8.append(np.array([x,y,z]))
                    
    Answer_1 = np.array(Answer_1)
    Answer_3 = np.array(Answer_3)
    Answer_7 = np.array(Answer_7)
    Answer_8 = np.array(Answer_8)
    
    return Answer_8,Answer_7,Answer_1,Answer_3,side

def plot3dim(n,fig,ax,PLOT = True,given = 'b'):
    Answer_8,Answer_7,Answer_1,Answer_3,side = find_ans(n)
    
    if PLOT == True:
        ax.scatter(Answer_1[:,0], Answer_1[:,1], Answer_1[:,2], zdir='z', s=10,color = 'g')
        plt.hold(True)
        ax.scatter(Answer_3[:,0], Answer_3[:,1], Answer_3[:,2], zdir='z', s=10,color = 'r')
        ax.scatter(Answer_7[:,0], Answer_7[:,1], Answer_7[:,2], zdir='z', s=10,color = 'y')
        ax.scatter(Answer_8[:,0], Answer_8[:,1], Answer_8[:,2], zdir='z', s=10,color = 'k')
        ax.set_xlim3d(0,1)
        ax.set_ylim3d(0,1)
        ax.set_zlim3d(0,1)
#         if index == 0:
#             s = '(0,0,0)'
#         elif index == 1:
#             s = '(0,0,1)'
#         elif index == 2:
#             s = '(0,1,0)'
#         elif index == 3:
#             s = '(1,0,0)'
#         elif index == 4:
#             s = '(0,1,1)'
#         elif index == 5:
#             s = '(1,0,1)'
#         elif index == 6:
#             s = '(1,1,0)'
#         elif index == 7:
#             s = '(1,1,1)'

#         if index != 8:
#             ax.set_title("Points closest to the point " + s)
#         else:
#             ax.set_title("Points closest to the center")

        fig.show()
        plt.hold(True)
        
#     Vol = Answer.shape[0] * (side**3)
#     print("Volume = ",Vol)
#     return Vol
```
