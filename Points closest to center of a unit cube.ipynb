```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time
```

```python
A1 = np.array([0,0,0])
A2 = np.array([0,0,1])
A3 = np.array([0,1,0])
A4 = np.array([1,0,0])
A5 = np.array([0,1,1])
A6 = np.array([1,0,1])
A7 = np.array([1,1,0])
A8 = np.array([1,1,1])
A9 = np.array([0.5,0.5,0.5])
A = np.array([A1,A2,A3,A4,A5,A6,A7,A8,A9])
```

```python
def initialize(n):
    X = np.linspace(0,1,n)
    Y = np.linspace(0,1,n)
    Z = np.linspace(0,1,n)
    side = X[1] - X[0]
    return (X,Y,Z, side)

def nearest_point(a,index):
    d = []
    for i in A:
        d.append(np.linalg.norm(a - i))
    m = np.argmin(d)
    if m == index:
        return 1
    return 0

def find_ans(n,index):
    Answer = []
    X,Y,Z,side = initialize(n)
    for i in X:
        for j in Y:
            for k in Z:
                x = i + side/2
                y = j + side/2
                z = k + side/2
                res = nearest_point(np.array([x,y,z]),index)
                if res==1:
                    Answer.append(np.array([x,y,z]))
    Answer = np.array(Answer)
    return Answer,side

def plot3dim(n,index,PLOT = True,given = 'b'):
    if index>8 or index<0:
        print("Invalid index")
        return
    
    Answer,side = find_ans(n,index)
    
    if PLOT == True:
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter(Answer[:,0], Answer[:,1], Answer[:,2], zdir='z', s=10,color = given)
        ax.set_xlim3d(0,1)
        ax.set_ylim3d(0,1)
        ax.set_zlim3d(0,1)
        if index == 0:
            s = '(0,0,0)'
        elif index == 1:
            s = '(0,0,1)'
        elif index == 2:
            s = '(0,1,0)'
        elif index == 3:
            s = '(1,0,0)'
        elif index == 4:
            s = '(0,1,1)'
        elif index == 5:
            s = '(1,0,1)'
        elif index == 6:
            s = '(1,1,0)'
        elif index == 7:
            s = '(1,1,1)'

        if index != 8:
            ax.set_title("Points closest to the point " + s)
        else:
            ax.set_title("Points closest to the center")

        fig.show()
        plt.hold(True)
        
    Vol = Answer.shape[0] * (side**3)
    print("Volume = ",Vol)
    return Vol
```
